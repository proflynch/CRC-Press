# Plotting with MatPLotLib

```python
# A simple plot.
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2 , 2 , 100)
y = x**2
plt.xlabel("x")
plt.ylabel("y")
plt.title("$y=x^2$")
plt.plot(x , y)
plt.savefig("Parabola.png" , dpi = 400)
plt.show()
```

```python
# Parametric plot.
import numpy as np
import matplotlib.pyplot as plt
t=np.linspace(0 , np.pi , 1000)
x = 0.7 * np.sin(t + 1) * np.sin(5 * t)
y = 0.7 * np.cos(t + 1) * np.sin(5 * t)
plt.plot(x , y , color = "magenta")
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.title("Parametric Plot")
plt.show()
```

```python
# Two curves in one plot.
# Add a grid so it looks like oscilloscope output.
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(0 , 2 , 0.01)
c = 1 + np.cos(2 * np.pi * t)
s = 1 + np.sin(2 * np.pi * t)
plt.plot(t , s , 'g--' , t , c , 'm-.')
plt.xlabel("Time (s)")
plt.ylabel('Voltage (V)')
plt.title('Voltage-Time Plot')
plt.grid(True)
# Save the plot as png. 
plt.savefig('Voltage-Time Plot.png')
plt.show()
```

```python
# Subplots and legends.
import matplotlib.pyplot as plt
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t=np.arange(0.0, 5.0, 0.1)
plt.figure(1)
plt.subplot(211)  #subplot(num rows,num cols,fig num)
plt.plot(t,f(t),'bo',t,f(t),'k',label='damping')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (cm)')
plt.title('Damped pendulum')
legend = plt.legend(loc='upper center',shadow=True)
plt.subplot(212)
plt.plot(t, np.cos(2*np.pi*t),'g--',linewidth=4)
plt.xlabel('time (s)')
plt.ylabel('Amplitude (cm)')
plt.title('Undamped pendulum')
plt.subplots_adjust(hspace=0.8)
plt.show()
```

```python
# Load data from an excel spreadsheet and fit polynomial curve.
# In Google Colab, drag and drop the data file on the left.
# In Spyder, the data file has to be in the same folder as the program.
import pandas as pd  # Data Science.
import numpy as np
import matplotlib.pyplot as plt
mydata=pd.read_excel('Kam_Data_nA.xlsx',sheet_name='Sheet1')
df=pd.DataFrame(mydata)
xs=(df['Volt'])
ys=(df['nAJ'])
z=np.polyfit(xs , ys , 10) # Polynomial of degree 10.
p=np.poly1d(z)
print('polynomial= ', p)
plt.axis([-0.1 , 0.8 , -100 , 300])
plt.xlabel('V (volts)', fontsize = 15)
plt.ylabel('I (nA)', fontsize = 15)
plt.title('Data Points', fontsize = 15)
plt.plot(xs,ys,'r+')
plt.show()
```
